```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

mortality_1854_55  = pd.read_excel('Downloads/mini project/mini project/Nightingale/raw_night_1854_1855.xlsx')
mortality_1855_56 = pd.read_excel('Downloads/mini project/mini project/Nightingale/raw_night_1855_1856.xlsx')
```

```python
# . DATA EXPLORATION
print("Data Exploration: 1854–1855")
print(mortality_1854_55.head())
print(mortality_1854_55.info())
print(mortality_1854_55.describe())
```

```python
# DATA EXPLORATION: 1855–1856
print("\nData Exploration: 1855–1856")
print(mortality_1855_56.head())
print(mortality_1855_56.info())
print(mortality_1855_56.describe())
```

```python
# Data c

print("\nMissing values (1854–1855):")
print(mortality_1854_55.isna().sum())
mortality_1854_55 = mortality_1854_55.dropna()
```

```python
print("\nMissing values (1855–1856):")
print(mortality_1855_56.isna().sum())
mortality_1855_56 = mortality_1855_56.dropna()
```

```python
# COXCOMB PLOT: 1854–1855

N1 = len(mortality_1854_55)
theta1 = np.linspace(0.0, 2 * np.pi, N1, endpoint=False)
width1 = (2 * np.pi) / N1

disease1 = mortality_1854_55['Disease Deaths'].values
wounds1 = mortality_1854_55['Wounds'].values
other1 = mortality_1854_55['Other Causes'].values

fig1, ax1 = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
# Plot stacked bars
ax1.bar(theta1, disease1, width=width1, label='Disease', color='navy', alpha=0.8)
ax1.bar(theta1, wounds1, width=width1, bottom=disease1, label='Wounds', color='red', alpha=0.8)
ax1.bar(theta1, other1, width=width1, bottom=disease1+wounds1, label='Other Causes', color='gray', alpha=0.8)

# Customize appearance
ax1.set_xticks(theta1)
ax1.set_xticklabels(mortality_1854_55['Month'], rotation=90)
ax1.set_title("Causes of Death (1854–1855)", fontsize=14, fontweight='bold')
ax1.grid(False)  # remove circular grid lines
ax1.set_frame_on(False)  # remove outer polar frame
ax1.legend(loc='upper right')

plt.show()
```

The Coxcomb plot for 1854–1855 shows that the majority of soldier deaths were caused by **disease**, with smaller numbers due to **wounds** and **other causes**. Disease deaths increased dramatically from June to August, despite relatively stable army sizes. This highlights the poor sanitary conditions and prevalence of preventable illnesses in military hospitals during this period.

```python
# COXCOMB PLOT: 1855–1856

N2 = len(mortality_1855_56)
theta2 = np.linspace(0.0, 2 * np.pi, N2, endpoint=False)
width2 = (2 * np.pi) / N2

disease2 = mortality_1855_56['Disease Deaths'].values
wounds2 = mortality_1855_56['Wounds'].values
other2 = mortality_1855_56['Other Causes'].values

fig2, ax2 = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
# Plot stacked bars
ax2.bar(theta2, disease2, width=width2, label='Disease', color='navy', alpha=0.8)
ax2.bar(theta2, wounds2, width=width2, bottom=disease2, label='Wounds', color='red', alpha=0.8)
ax2.bar(theta2, other2, width=width2, bottom=disease2+wounds2, label='Other Causes', color='gray', alpha=0.8)

# Customize appearance
ax2.set_xticks(theta2)
ax2.set_xticklabels(mortality_1855_56['Month'], rotation=90)
ax2.set_title("Causes of Death (1855–1856)", fontsize=14, fontweight='bold')
ax2.grid(False)  # remove circular grid lines
ax2.set_frame_on(False)  # remove outer polar frame
ax2.legend(loc='upper right')

plt.show()
```

The Coxcomb plot for 1855–1856 shows that disease remained the leading cause of death, though the total numbers are lower compared to the previous year. Deaths from wounds and other causes remained relatively low and consistent. This suggests that improvements in hygiene may have started to reduce mortality, The overall decline in deaths demonstrates the positive effect of improved hygiene and medical care, showing the impact of Florence Nightingale's efforts.
